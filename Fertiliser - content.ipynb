```python
crop_name = input()
N = int(input("Enter Nitrogen value : "))
P = int(input("Enter Phosphorus value : "))
K = int(input("Enter Potassium value : "))
PH=int(input("Enter the pH value : "))

n_actual = data[data['Crop'] == crop_name]['N'].iloc[0]
p_actual = data[data['Crop'] == crop_name]['P'].iloc[0]
k_actual = data[data['Crop'] == crop_name]['K'].iloc[0]
p_actual=  data[data['Crop'] == crop_name]['pH'].iloc[0]

n = n_actual - N
p = p_actual - P
k = k_actual - K
ph=p_actual- PH
'''temp = {abs(n): "N", abs(p): "P", abs(k): "K"}
max_value = temp[max(temp.keys())]'''
nkey=''
pkey=''
kkey=''
phkey=''
if n < -10:
    nkey = 'NHigh'
elif n>10:
    nkey = "Nlow"
else:
    nkey="Nok"
    
if p < -10:
    pkey = 'PHigh'
elif p>10:
    pkey = "Plow"
else:
    pkey="Pok"

if k < -10:
    kkey = 'KHigh'
elif k>10:
    kkey = "Klow"
else:
    kkey="Kok"
    
if ph<-10:
    phkey='phHigh'
elif ph>10:
    phkey='phLow'
else:
    phkey='phok'
```
